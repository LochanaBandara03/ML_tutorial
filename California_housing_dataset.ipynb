```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

#Get California Housing Dataset

from sklearn.datasets import fetch_california_housing
housing =fetch_california_housing()
housing;
```

```python
housing_df = pd.DataFrame(housing["data"], columns=housing['feature_names'])
housing_df["target"] = pd.Series(housing["target"])
housing_df.head()
```

```python
#How many samples
len(housing_df)
```

```python
#Import the ridge model class
from sklearn.linear_model import Ridge

np.random.seed(42)

#Splits into X(features) and y(lables)
X = housing_df.drop("target", axis=1)
y = housing_df["target"]

#Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#Fit the model
model = Ridge()
model.fit(X_test, y_test)

#Score of the model
model.score(X_test, y_test)
```

```python
#Import the RandomForestClassifer model
from sklearn.ensemble import RandomForestRegressor

np.random.seed(42)

#Split the data into X(features) and y(lables)
X= housing_df.drop("target", axis=1)
y= housing_df["target"]

#Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#Fit the model
model = RandomForestRegressor()
model.fit(X_train, y_train)

#Score the model
model.score(X_test, y_test)
```

We get a good boost in score on the test set by changing the model.

This is another incredibly important concept in machine learning, if at first something doesn't achieve what you'd like, experiment, experiment, experiment!
