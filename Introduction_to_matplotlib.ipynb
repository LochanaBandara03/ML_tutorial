##A quick Matplotlib Workflow

* Import matplotlib - For example, import matplotlib.pyplot as plt).
* Prepare data - This may be from an existing dataset (data analysis) or from the outputs of a machine learning model (data science).
* Setup the plot - In other words, create the Figure and various Axes.
* Plot data to the Axes - Send the relevant data to the target Axes.
* Cutomize the plot - Add a title, decorate the colours, label each Axis.
* Save (optional) and show - See what your masterpiece looks like and save it to file if necessary.


```python
# A matlplotlib workflow

#0. Import and get matplot ready
# %matplotlib inline #Not necessary
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

#2. Setup plot (Figure and Axes)
fig, ax = plt.subplots(figsize=(10,10))

#3. Plot data
ax.plot(x, y)

#4. Customiize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. Save & show
fig.savefig("/content/sample_data/simple-plot.png")
```

###Creating Figures with multiple Axes with Subplots

Option 01

```python
#Creating 4 subplots - each axes have own variable name
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))

#Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

Option 02

```python
# Creating 4 subplots - subplots with single ax variable
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Indexing to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.rand(1000));
```

###Plotting more advanced plots

```python
#Data analysis on patients on over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
#Create a scatter plot from that
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target", #color the dot by target value
             figsize=(10,6));
```

```python
#Create a figure and axes instance
fig, ax = plt.subplots(figsize=(10,6))

#Plot data from the DataFrame
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#Customize the x-axis limits
ax.set_xlim([45, 100]);
```

We can create more comprehensive plot

```python
#Create figure and axes
fig, ax =  plt.subplots(figsize=(10,6))

#plot data directly to the axes
scatter = ax.scatter(over_50["age"],
                     over_50["chol"],
                     c= over_50["target"])

#Customize the plot parameters
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#Setup the legend
ax.legend(*scatter.legend_elements(),
          title="Target");

#Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

###Plotting multiple plots on the same figure

```python
#Setup columns
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True, #Use same x-axis
                               figsize=(10,8))

#...............Axis 0 : Heart Disease and Cholesterol Levels...............

#plot data directly to the ax0
scatter = ax0.scatter(over_50["age"],
                     over_50["chol"],
                     c= over_50["target"])

#Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(), title="Target")

#Setup the legend
ax0.legend(*scatter.legend_elements(),
          title="Target");

#Add a meanline
ax0.axhline(over_50["chol"].mean(),
           color='b',
           linestyle="--",
           label="Average");

#...............Axis 1 : Heart Disease and Max Heart Rate Levels...............

#plot data directly to the ax1
scatter = ax1.scatter(over_50["age"],
                     over_50["thalach"],
                     c= over_50["target"])

#Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart rate levels");
ax1.legend(*scatter.legend_elements(), title="Target")

#Setup the legend
ax1.legend(*scatter.legend_elements(),
          title="Target");

#Add a meanline
ax1.axhline(over_50["thalach"].mean(),
           color='b',
           linestyle="--",
           label="Average");
```
